```python
df.corr()
```

```python
index_vals = df['Genre'].astype('category').cat.codes

fig = go.Figure(data=go.Splom(
                dimensions=[dict(label='User Rating',
                                 values=df['User Rating']),
                            dict(label='Reviews',
                                 values=df['Reviews']),
                            dict(label='Price',
                                 values=df['Price'])],
                showupperhalf=False, 
                text=df['Name'],
                marker=dict(color=index_vals,
                            showscale=False, # colors encode categorical variables
                            line_color='white', line_width=0.5)
                ))


fig.update_layout(
    title='Books',
    width=1000,
    height=1000,
)

fig.show()
```

not enough correlation spotted on our report we can see the relations between variables

```python
genre_by_year = df.groupby('Year')['Genre'].value_counts().reset_index(level=0).rename(columns={'Genre': 'Genre count'}, index={'index': 'Genre'})
genre_by_year
```

```python
fig = px.line(genre_by_year, x='Year', y='Genre count', color= genre_by_year.index, title='Movies By Genre in Each Year')
fig.show()
```

Fiction book dropped high rated selling from 2014, Non Fiction books bounced back

```python
df
rating_by_year = df.groupby('Year')['User Rating'].value_counts().reset_index(level=0).rename(columns={'User Rating': 'Rating count'}, index={'index': 'User Rating'})
rating_by_year.head(20)
```

```python
fig = px.line(rating_by_year, x='Year', y='Rating count', color= rating_by_year.index, title='Books By Rating in Each Year')
fig.show()
```

4.8 rating increased in 2018

```python
fig = px.scatter(df, x='Year', y='Price', title='Price of the Books in Each Year', hover_data = df[['Name','Author']])
fig.show()
```

```python
fig = px.scatter(df, x='Year', y='Reviews', title='Number of Reviews in Each Year', color='Genre',hover_data = df[['Name','Author']])
fig.show()
```

```python
fig = px.scatter(df, x='Year', y='User Rating', title='User Rating in Each Year',color='Genre', hover_data = df[['Name','Author']])
fig.show()
```

User rating is done uniformly for each year

```python
top_20 = df.sort_values('User Rating', ascending=False)[:20]
top_20
```

```python
fig = px.bar(top_20, x='Name', y= 'User Rating',  hover_data = top_20[['Year','Genre', 'Price']], color='Genre')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

From top 20 books we see that Fiction books have most user ratings

```python

bottom_20 = df.sort_values('User Rating')[:20]
bottom_20
```

```python
fig = px.bar(bottom_20, x='Name', y= 'User Rating',  hover_data = bottom_20[['Year','Genre', 'Price']], color='Genre')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

The opposite case is true for bottom 20 rated books. Less ratings are given for fiction books. interesting find!

```python
#Most 20 reviewed books

top_20_reviews = df.sort_values('Reviews', ascending=False)[:20]
top_20_reviews
```

```python
fig = px.bar(top_20_reviews, x='Name', y= 'Reviews',  hover_data = top_20_reviews[['Year','Genre', 'Price']], color='Genre')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

Fiction books are most reviewed! The Alchemist is one of my fav book! :D
